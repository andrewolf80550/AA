## COMPLETAR LOS DATOS DE ACUERDO A LAS PALABRAS CLAVES DE LA DESCRIPCIÓN DEL SUSESO

```python
dic = {'MANTO': ['falla','indicación','funcionamiento', 'indicacion', 'motor', 'mantenimiento', 'sistema','fail','vibraci','prende','la luz','tren de aterrizaje','fisura','RPM','inoperativo','anotaci','desinfla','olor','chip','velocimetro','fisura','escape','sobretorque','encendido','daño','humo','despresuriza','alabeo','asimetria','pierde','potencia','suelta','pincha','anormal','generadores','intermitentemente','oxigeno','ruptura','combustible','desfase','enciende','fuga','MANIFOLD PRESSURE','presi'],
      'GAAP': ['golpe','colisión','colision','impacto','poste'],
       'RIPP': ['incursión', 'incursion','ingreso','ingreso','carro','pista','calle de rodaje','no autorizado'],
        'CFIT-ALAR': ['vuelo controlado','terreno','terrain','alarma'],
        'MACA': ['resolution advisory','tcas','dos aeronaves'],
       'FUEGO ENEMIGO':['bala', 'proyectil', 'disparo', 'orificio','ataque','terrorista','impactada'],
        'MERCANCIAS PELIGROSAS': ['derrame','residuos'],
        'ARMAMENTO': ['armamento', 'bomba', 'carchuchos', 'arma']}
```

```python
for k in dic:
    for p in dic[k]:
       # data.loc[p in data['DESCRIPCION_EVESO'], 'SUBTIPO_EVESO'] = k #filtro por la palabra
        data['DESCRIPCION_EVESO'].filter(like= p)
        # print(data['DESCRIPCION_EVESO'].filter(like= p, axis=0))
        data['SUBTIPO_EVESO'] = np.where(data['DESCRIPCION_EVESO'].str.contains( p, case=False, na=False),k, data['SUBTIPO_EVESO'])
```
